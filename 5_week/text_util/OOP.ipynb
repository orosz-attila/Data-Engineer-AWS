```python
import nltk
```

## Version 1. 

```python
class TextUtil():

    def __init__(self, filename):
        "Represents a text"
        self.filename = filename

    def read_text(self):
        "Reads the text from file"
        file = open(self.filename, "r")
        self.text = file.read().replace('\n', ' ')
        return self.text
  
    def print_raw(self, text):
        self.__print_text = f'raw text is: {self.text}'
        return self.__print_text
     
    def word_split (self, text):
        "Splits the text to words (list)"
        words = nltk.word_tokenize(text)
        self.word_list = [word.lower() for word in words if word.isalpha()]
        return self.word_list 
    
    def one_word_count (self, word_list, word):
        "Counts one word given as input"
        count = word_list.count(word)
        self.dict_1 = {word: count}
        return self.dict_1
    
    def all_words_count (self, word_list):
        "Counts all words in a list given as input"
        self.dict_all = {}
        for word in word_list:
             self.dict_all[word]=word_list.count(word)
        return self.dict_all
    
    def save_file (self, text, dict_all):
        "Saves the dictionary with word/count as key/value pairs"
        with open(f'{text}.txt', 'w', encoding = 'utf-8') as file:
            for key, value in dict_all.items(): 
                file.write("%s %s\n" % (key, value))
```

```python
# initialize loreley text file
loreley = TextUtil("heine_loreley.txt")
```

```python
# read loreley text 
loreley.read_text()
```

```python
# split loreley text into words 
loreley.word_split(loreley.text)
```

```python
# count 1 given word in loreley text
loreley.one_word_count(loreley.word_list, 'ich')
```

```python
# count all words in loreley text
loreley.all_words_count(loreley.word_list)
```

```python
# saving the file 
loreley.save_file('loreley', loreley.dict_all)
```

## Version 2. 

```python
class TextUtil():

    def __init__(self, file_name):
        "Represents a text"
        self.file_name = file_name
        self.read_text()
        self.word_list()
        self.word_count()

    def read_text(self):
        "Reads the text from file"
        with open(self.file_name, 'r') as file:
            self.text = file.read().replace('\n', ' ')
            
    def word_list (self):
        "Splits the text to words (list)"
        words = nltk.word_tokenize(self.text)
        self.word_list = [word.lower() for word in words if word.isalpha()]
    
    def word_count (self):
        "Counts all words"
        self.word_count = {}
        for word in self.word_list:
             self.word_count[word] = self.word_list.count(word)
                
    def save_file (self):
        "Saves the dictionary with word/count as key/value pairs"
        with open('word_count' + '_' f'{self.file_name}', 'w', encoding = 'utf-8') as file:
            for key, value in self.word_count.items(): 
                file.write("%s %s\n" % (key, value))
```
